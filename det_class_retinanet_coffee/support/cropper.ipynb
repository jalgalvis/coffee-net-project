```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "y",
          "axes.facecolor" : "None",
          "text.color" : 'w'}
plt.rcParams.update(params)
```

```python
H_IMGS = 12
V_IMGS = 9
SIZE_IMG = 224
```

```python
# the following way colors change:
# import cv2
# im = cv2.imread("data/test_image/coffee_sheet/coffee.jpg")

# this way image is more similar than original
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    directory= 'data/test_image/',
    color_mode='rgb',
    image_size=(V_IMGS * SIZE_IMG, H_IMGS * SIZE_IMG),
)
```

```python
list_dataset = list(train_ds.as_numpy_iterator())
im = np.array(list_dataset[0][0][0])
```

```python
arrayIm = np.empty((0,SIZE_IMG,SIZE_IMG,3))
count = 0
for i in range(0,H_IMGS * SIZE_IMG, SIZE_IMG):
    for j in range(0,V_IMGS * SIZE_IMG,SIZE_IMG):
        new = im[j:j+SIZE_IMG, i:i+SIZE_IMG, :]
        arrayIm = np.vstack((arrayIm, new[np.newaxis,:,:,:]))
        count += 1

```

```python
plt.figure(figsize=(10, 10))
i = 0
n = np.random.randint(0,count,(9))
for images in arrayIm[n,:,:,:]:
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images.astype("uint8"))
    plt.axis("off")
    i += 1
```
