```
start_time = time.time()

df = pd.read_csv('annotations/all_labels.csv')

images_directory = 'images/'
total_files = 0

for filename in os.listdir(images_directory):
  file_number = 0
  df1 = df[df['filename']==filename]
  im = Image.open(fp=images_directory + filename)
  im = np.asarray(im)
  if df1['class'].iloc[0] == 'good':
    directory = '/content/'
  else:
    directory = '/content/'

  for xmin, xmax, ymin, ymax in zip(df1['xmin'], df1['xmax'], df1['ymin'], df1['ymax']):
    out = Image.fromarray(np.uint8(im[ymin:ymax, xmin:xmax]))
    out.save(fp=f'{directory}{os.path.splitext(filename)[0]}-{file_number}.jpg')
    file_number += 1
    total_files += 1

  print(f'processing {filename}, total saved files: {total_files}')

end_time = time.time()
elapsed_time = end_time - start_time
print(f'Elapsed time: {int(elapsed_time/60)}m')
```
